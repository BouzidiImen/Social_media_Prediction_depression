# Test best Models:
## BiLSTM+ GloVe:
    10k not preprocessed Tweet:

```python
import pandas as pd
test_data = pd.read_csv("data/test_data.csv")
```

```python
classes_index = {
    1 : "Depressed",
    0 : "Not depressed"
}
from tensorflow import keras
import pickle
model = keras.models.load_model('Exp_10k_32ep/model_BiGloVe.h5')
# loading
with open('Exp_10k_32ep/tokenizer_BiGlove.pickle', 'rb') as handle:
    tokenizer = pickle.load(handle)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
def predict_tweet_BiGlove(text_to_predict):
    test_sequence = tokenizer.texts_to_sequences([text_to_predict])
    test_data = pad_sequences(test_sequence, maxlen=99)
    predictions = model.predict(test_data)
    prediction = np.argmax(predictions)
    print("The person who tweeted this is {}\n".format(classes_index[prediction]))
    ranking = predictions.argsort()[-2:][::-1]
    for i in reversed(range(0, 2)):
        rank = ranking[0][i]
        print("{:>10} : {:.5f}%".format(classes_index[rank], predictions[0][rank]*100))
```

```python
for i in range(0,40):
    print(test_data["Tweet"][i])
    print()
    predict_tweet_BiGlove(test_data["Tweet"][i])
    print("*************************************************************************************")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np 
from tqdm import trange
from tensorflow.keras.utils import to_categorical
def evaluate_test(data):
    d=data["Tweet"].tolist()  
    sequences = tokenizer.texts_to_sequences(d)
    seqs = pad_sequences(sequences, maxlen=99)
    y = to_categorical(data["label"].tolist())
    score = model.evaluate(seqs,y, verbose=2)
    print('Test accuracy:', score[ 1 ])
    Y_pred = model.predict(seqs)
    y_pred = np.argmax(Y_pred, axis=1)
    y_classes = np.argmax(y, axis=1)
    target_names = [ 'Depressed', 'Not depressed' ]
    print(confusion_matrix(y_classes, y_pred) )
    print(classification_report(y_classes, y_pred, target_names=target_names))
```

```python
%%time
evaluate_test(test_data)
```

## Preprocessed 10k Tweet 32 batch size:

```python
import pandas as pd
data = pd.read_csv("Exp_prep_10k_32ep/test_data.csv")
```

```python
classes_index = {
    1 : "Depressed",
    0 : "Not depressed"
}
from tensorflow import keras
import pickle
model = keras.models.load_model('Exp_prep_10k_32ep/model_BiGloVe.h5')
# loading
with open('Exp_prep_10k_32ep/tokenizer_BiGlove.pickle', 'rb') as handle:
    tokenizer = pickle.load(handle)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
def predict_tweet_BiGlove(text_to_predict):
    test_sequence = tokenizer.texts_to_sequences([text_to_predict])
    test_data = pad_sequences(test_sequence, maxlen=197)
    predictions = model.predict(test_data)
    prediction = np.argmax(predictions)
    print("The person who tweeted this is {}\n".format(classes_index[prediction]))
    ranking = predictions.argsort()[-2:][::-1]
    for i in reversed(range(0, 2)):
        rank = ranking[0][i]
        print("{:>10} : {:.5f}%".format(classes_index[rank], predictions[0][rank]*100))
```

```python
for i in range(0,20):
    print(data["Tweet"][i])
    print()
    
    predict_tweet_BiGlove(data["Tweet"][i])
    print("*************************************************************************************")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np 
from tqdm import trange
from tensorflow.keras.utils import to_categorical
def evaluate_test(data):
    d=data["Tweet"].astype(str).tolist()    
    sequences = tokenizer.texts_to_sequences(d)
    seqs = pad_sequences(sequences, maxlen=197)
    y = to_categorical(data["label"].tolist())
    score = model.evaluate(seqs,y, verbose=2)
    print('Test accuracy:', score[ 1 ])
    Y_pred = model.predict(seqs)
    y_pred = np.argmax(Y_pred, axis=1)
    y_classes = np.argmax(y, axis=1)
    target_names = [ 'Depressed', 'Not depressed' ]
    print(confusion_matrix(y_classes, y_pred) )
    print(classification_report(y_classes, y_pred, target_names=target_names))
```

```python
%%time
evaluate_test(data)
```
